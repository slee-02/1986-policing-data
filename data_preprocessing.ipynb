### 성별 (RPTER_SEX)

```python
def sex_sub(n):
    if n == 1:
        return '남성'
    elif n == 2:
        return '여성'
    else:
        return '불상'
```

```python
df['RPTER_SEX'] = df['RPTER_SEX'].map(sex_sub)
```

```python
df['RPTER_SEX'].value_counts()
```

## 파생변수

```python
df["Year"] = df["접수완료일시"].dt.year
df["Month"] = df["접수완료일시"].dt.month
df['Year Month'] = df['접수완료일시'].map(lambda x : str(x)[:7])
df["Day of Week"] = df["접수완료일시"].dt.day_name()
df["Hour"] = df["접수완료일시"].dt.hour
df['일자'] = df['접수완료일시'].dt.date
```

```python
df.head()
```

```python
df = df.rename(columns={'Year' : '연도', 'Month' : '월', 'Year Month' : '연월', 'Day of Week' : '요일', 'Hour' : '시간대'})
```

```python
df
```

```python
df = df.sort_values('접수완료일시')
```

```python
df['접수부서'].value_counts()
```

```python
def dept_replace(x):
    if '대전' in x:
        return '대전청'
    elif '세종' in x:
        return '세종청'
    else:
        return '충남청'
```

```python
df['접수부서'] = df['접수부서'].map(dept_replace)
```

```python
df.head()
```

```python
df['접수부서'].value_counts()
```

```python
dayofweek_dict = {'Monday' : '월',
                  'Tuesday' : '화',
                  'Wednesday' : '수',
                  'Thursday' : '목',
                  'Friday' : '금',
                  'Saturday' : '토',
                  'Sunday' : '일'
                  }
```

```python
df['요일'] = df['요일'].map(dayofweek_dict)
```

```python
df['요일'].value_counts()
```
