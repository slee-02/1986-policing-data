```python
import pandas as pd
import numpy as np
import json

# 시각화 라이브러리
import seaborn as sns
import plotly.express as px
import matplotlib.pyplot as plt
# import folium
# from folium.plugins import MarkerCluster
```

```python
file_path = './data/'
```

```python
df_dj = pd.read_excel(f'{file_path}daejeon_vp_percent.xlsx')
df_cn = pd.read_excel(f'{file_path}chungnam_vp_percent.xlsx')
df_sj = pd.read_excel(f'{file_path}sejong_vp_percent.xlsx')
```

# 시각화

## 대전광역시

```python
plt.figure(figsize=(15, 5));

plt.subplot(1, 3, 1)
sns.barplot(data=df_dj, x='지역별', y='보이스피싱 피해 비율', hue='연도');
plt.title('대전광역시 행정구별 인구수 대비 보이스피싱 피해비율', fontsize = 10);

plt.subplot(1, 3, 2)
sns.barplot(data=df_dj, x='지역별', y='보이스피싱 건수', hue='연도');
plt.title('대전광역시 행정구별 보이스피싱 신고건수', fontsize = 10);

plt.subplot(1, 3, 3)
sns.barplot(data=df_dj, x='지역별', y='인구', hue='연도');
plt.title('대전광역시 행정구별 인구수', fontsize = 10);
```

대전광역시의 경우 인구수의 변화가 거의 없는 것에 반해 실제 보이스피싱 건수가 감소하므로서 대전광역시의 인구수대비 피해비율이 감소하였다.

## 충청남도

```python
plt.figure(figsize=(10, 15));

plt.subplot(3, 1, 1)
sns.barplot(data=df_cn, x='지역별', y='보이스피싱 피해 비율', hue='연도');
plt.title('충청남도 행정시군구별 인구수 대비 보이스피싱 피해비율', fontsize = 10);

plt.subplot(3, 1, 2)
sns.barplot(data=df_cn, x='지역별', y='보이스피싱 건수', hue='연도');
plt.title('충청남도 행정시군구별 보이스피싱 신고건수', fontsize = 10);

plt.subplot(3, 1, 3)
sns.barplot(data=df_cn, x='지역별', y='인구', hue='연도');
plt.title('충청남도 행정시군구별 인구수', fontsize = 10);
```
