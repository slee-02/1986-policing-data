```python
import pandas as pd
import numpy as np
```

```python
file_path = './data/'
```

# 대전광역시 인구 데이터 병합

```python
daedeog = pd.read_csv(f'{file_path}daejeon/daedeog_gu.csv', encoding="cp949")
dong = pd.read_csv(f'{file_path}daejeon/dong_gu.csv', encoding="cp949")
jung = pd.read_csv(f'{file_path}daejeon/jung_gu.csv', encoding="cp949")
seo = pd.read_csv(f'{file_path}daejeon/seo_gu.csv', encoding="cp949")
yuseong = pd.read_csv(f'{file_path}daejeon/yuseong_gu.csv', encoding="cp949")
```

```python
display(daedeog, dong, jung, seo, yuseong)
```

```python
df_daejeon = pd.concat([daedeog, dong, jung, seo, yuseong])
df_daejeon
```

```python
df_daejeon = df_daejeon.reset_index().drop(columns='index')
df_daejeon
```

```python
dj_sigungu_list = []
for idx in list(df_daejeon.index):
    a = df_daejeon['지역별(1)'][idx].split(' ')[1]
    dj_sigungu_list.append(a)
```

```python
df_daejeon['지역별'] = dj_sigungu_list
```

```python
df_daejeon = df_daejeon[['지역별', '2020', '2021']]
df_daejeon
```

```python
df_daejeon.to_csv(f'{file_path}daejeon_population.csv', index=False)
```

# 충청남도 인구 데이터 병합

```python
gyeryong = pd.read_csv(f'{file_path}chungnam/gyeryong_si.csv', encoding="cp949")
dangjin = pd.read_csv(f'{file_path}chungnam/dangjin_si.csv', encoding="cp949")
cheongyang = pd.read_csv(f'{file_path}chungnam/cheongyang_gun.csv', encoding="cp949")
nonsan = pd.read_csv(f'{file_path}chungnam/nonsan_si.csv', encoding="cp949")
yesan = pd.read_csv(f'{file_path}chungnam/yesan_gun.csv', encoding="cp949")
hongseong = pd.read_csv(f'{file_path}chungnam/hongseong_gun.csv', encoding="cp949")
boryeong = pd.read_csv(f'{file_path}chungnam/boryeong_si.csv', encoding="cp949")
taean = pd.read_csv(f'{file_path}chungnam/taean_gun.csv', encoding="cp949")
buyeo = pd.read_csv(f'{file_path}chungnam/buyeo_gun.csv', encoding="cp949")
cheonan = pd.read_csv(f'{file_path}chungnam/cheonan_si.csv', encoding="cp949")
gongju = pd.read_csv(f'{file_path}chungnam/gongju_si.csv', encoding="cp949")
geumsan = pd.read_csv(f'{file_path}chungnam/geumsan_gun.csv', encoding="cp949")
seosan = pd.read_csv(f'{file_path}chungnam/seosan_si.csv', encoding="cp949")
seocheon = pd.read_csv(f'{file_path}chungnam/seocheon_gun.csv', encoding="cp949")
ahsan = pd.read_csv(f'{file_path}chungnam/ahsan_si.csv', encoding="cp949")
```

```python
df_chungnam = df_chungnam.reset_index().drop(columns='index')
df_chungnam
```

```python
cn_sigungu_list = []
for idx in list(df_chungnam.index):
    a = df_chungnam['지역별(1)'][idx].split(' ')[1]
    cn_sigungu_list.append(a)
```

```python
df_chungnam['지역별'] = cn_sigungu_list
```

```python
df_chungnam = df_chungnam[['지역별', '2020', '2021']]
df_chungnam
```

```python
df_chungnam.to_csv(f'{file_path}chungnam_population.csv', index=False)
```
